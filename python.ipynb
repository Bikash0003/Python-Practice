```python
#classify number as positive, Negative and Zero
def classify_number(n):
  if n>0:
    return f"the given {(n)} is positive number"
  elif n<0:
    return f"The given number {(n)} is negative"
  else:
    return f"The given number {(n)} is null"
print (classify_number(2))
```

```python
def grading_system(n):
  if n>50:
    return "The student is just passed "
  elif n>70:
    return "The student is very good"
  else:
    return "The student is fail"
print(grading_system(30))
```

```python
# LOOPS
numbers = [5, 8, 2, 1]
total = 0
for num in numbers:
    total += num   # add each number to the running total
print("Total:", total)  # Outputs: Total: 16
```
