```python
def is_even(n):
  if n == 0:
    return f"The number {n} is neither even nor odd"
  elif n % 2 == 0:
    return True
  else:
    return False

print(is_even(0))
```

```python
data = {
    "Date": ["2021-01-01", "2021-01-02", "2021-01-03", "2021-01-04"],
    "Region": ["North", "South", "East", "West"],
    "Product": ["A", "B", "A", "C"],
    "Quantity": [1, 2, 3, 1],
    "Price": [10, 15, 10, 20]
}
df = pd.DataFrame(data)
print(df)
```

```python
# to calculate the quantity by grouping the region and product
```

```python
Potato =df.groupby(["Region","Product"])["Quantity"].sum().reset_index()
Potato
```

```python
Sales_per_region=df.groupby(['Region'])['Product'].count().reset_index()
Sales_per_region.rename (columns={ 'Product':'Transactions_count'})
```

```python
product_price_by_region=df.groupby(["Region","Product"]).agg(
    Transactional_count=("Product",'count'),
    Total_Price=("Price",'sum')
    ).reset_index()
product_price_by_region
```

```python
# Pivot Tables
pivoting=product_price_by_region.pivot(index="Region",columns="Product",values="Total_Price").fillna(0)
pivoting
```
